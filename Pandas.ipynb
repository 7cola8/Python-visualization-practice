## 数据排序

```python
data=pd.read_excel('Excel数据.xlsx',
              sheet_name=0 #读第几张表的数据,从0开始数
             )
```

```python
data.sort_values(by=['单价'],#排序对象
                ascending=False)#升序降序
```

```python
data.sort_values(by=['单价','数量（台）'],#排序对象，用多个去排可以防止数据相同并列的情况
                ascending=[False,True],#Ture升序，False降序
                inplace=True)#升序降序
```

```python
data
```
