## æŒ‰å€¼åˆå¹¶

```python
df1=pd.DataFrame({'first':[2,3,8],
                'second':[4,5,6]},
               index=['aa','aa','c'])

df2=pd.DataFrame({'second':[6,5,8],
                'third':[17,8,19]},
               index=['b','c','aa'])
```

```python
pd.merge(df1,df2)#é»˜è®¤æ±‚äº¤é›†
```

```python
pd.merge(df1,df2,how='outer')
#ç”¨howæ¥ä¿®æ”¹ outerä¿ç•™æ‰€æœ‰æ•°æ® innerä¿ç•™äº¤é›†ï¼ˆæ²¡ç©ºçš„ï¼‰ leftä¿ç•™å®Œæ•´1 rightä¿ç•™å®Œæ•´2
```

```python
#on='first' å¤šç›¸åŒåˆ—æ˜ï¼ŒæŒ‡å®šå…¶ä¸­ä¸€ä¸ª,1æ”¾1ä¸‹é¢ï¼Œå¤šçš„3æ”¾åé¢;ä¼šè‡ªåŠ¨è¡¥é½åç¼€
#suffixes=('')ä¿®æ”¹åˆ—æ˜åç¼€
```

```python
pd.merge(df1,df2,
         how='outer',
         left_on='first',
         right_on='second',
         suffixes=('_df1','_df2'),)
```

```python
pd.merge(df1,df2,
         how='outer',
         on='second',
         suffixes=('_df1','_df2'))
```

```python
data=pd.read_excel('Excelæ•°æ®.xlsx')
data
```

```python
for i in data.groupby(by=['åœ°åŒº']):#æŒ‰ç…§æŒ‡å®šåˆ—ï¼Œåˆ†æˆå°æ•°æ®
    print(i)
    print('*'*80)
```

```python
#å¤šé”®åˆ†ç»„
data.groupby(by=['åœ°åŒº','åŸå¸‚']).count()
```

```python
#å¤šé”®åˆ†ç»„
data.groupby(by=['åœ°åŒº','åŸå¸‚'])['é”€å”®äººå‘˜','é”€å”®é¢'].count()
#å¯ä»¥æ±‚å„ç§éœ€è¦çš„æ•°æ®å’Œå•Šå¹³å‡å•Šä»€ä¹ˆçš„ï¼Œå¾ˆæ–¹ä¾¿
```
