## 折线图

```python
#别忘了逗号!!!!!!!!!!超级容易忘记，但是每一个函数后面都要记得加！！！！！
import matplotlib.pyplot as plt
#调整画布
plt.rcParams['font.family'] = 'youyuan'
plt.rcParams['axes.unicode_minus'] = False
#能够支持中文的参数，可以自己调字体，自己去找
plt.figure(
    figsize=(5,3), #占几份像素,前面长，后面宽
    dpi=200,  #调整清晰度，即像素个数
    #还可以调颜色什么的，自己去学着玩
)
plt.plot(x,y,'r:*',color='#f65758',marker='>',label='第一条线')
# color调颜色，按运行顺序放最下可以调所有线条的颜色;maker点的形状;标签需要和关键词放一起
plt.plot(x,y1,'b-H',linewidth=1,linestyle='--',label='第二条线')
# linewidth是用来调整线的粗细;linestyle线的形状
#多条线的话最好分开写
#逗号后面加的是对线条的修改，再后面是加图例
plt.legend(loc=2,fontsize=7)#加入此函数才能显示图例label;内部第一个函数位置，第二个函数调整字体大小
plt.title('折线图',
          loc='center')
#标题位置，不支持中文，需要导入参数才可以;loc改变位置
#加坐标讲解
plt.xlabel('自变量',fontsize=15,loc='right')
plt.ylabel(r'$E_0=mc^2$')  # latex 公式
#修改刻度范围
plt.xlim(-0.5,9.5)
plt.ylim(0,1)
#修改xy轴标签
plt.xticks(
[i*0.5 for i in range(0,19)],#x轴数据
[i*0.5 for i in range(0,19)],#对应的标签，和数据的个数保持
)
#网格线
plt.grid(
linestyle='-.',
linewidth=0.5,
axis='y')
#文本信息
plt.text(x[4],y[4],'这里是一个测试文字')
#出一长串未运行的代码的话一般代表数据错误
plt.savefig('图片测试')
```

## 子图

### 基础子图示例

```python
Fig=plt.figure()
Fig.add_subplot(2,1,1)#行数，列数，第几个
#想在子图1绘图，一定在图一下面，图二上面，注意顺序
plt.plot(x,y)
plt.title('1')

Fig.add_subplot(2,1,2)
plt.bar(x,y)
plt.title('2')
#调整布局
plt.tight_layout()
```

### 子图组合

```python
Fig=plt.figure()
Fig.add_subplot(3,3,1)#行数，列数，第几个
Fig.add_subplot(3,3,(2,3))
Fig.add_subplot(3,3,(4,7))#最后括号修改占位
```

```python
Fig=plt.figure()
Fig.add_subplot(2,2,1)#行数，列数，第几个
plt.title('1')
plt.bar(x,y,label='a')
plt.legend()

Fig.add_subplot(2,2,2)
plt.pie(x)
plt.title('2')

Fig.add_subplot(2,1,2)
plt.plot(x,y)
plt.title('3')

#调整默认间距
plt.tight_layout()
```

```python
#还是推荐一个图下面写自己的形式，也好看一些
Fig=plt.figure()
zitu=Fig.add_subplot(2,2,2)#以变量形式命名后即使写在子图2后面也不会出错，容错率更高
zitu.plot(x,y)
zitu.set_title('2')
zitu.set_xlabel('自变量')

Fig.add_subplot(2,2,1)
plt.title('1')
plt.bar(x,y,label='a')
plt.legend()

plt.suptitle('大标题')
#调整默认间距
plt.tight_layout()
```

### 3D曲面图

```python
import numpy as np #构建数组，计算函数
ZZZ=np.arange(-5,5,0.01)
YY=np.arange(-5,5,0.01)
X,Y=np.meshgrid(ZZZ,YY)
Z=np.sin(X)+np.cos(Y)
```

```python
fig=plt.figure()
axl=fig.add_subplot(1,1,1,projection='3d')
axl.plot_surface(X,Y,Z,cmap='rainbow')#其实就是在平面图函数后面加3D
axl.set_xlabel('x')
axl.set_ylabel('y')
axl.set_zlabel('z')
```
